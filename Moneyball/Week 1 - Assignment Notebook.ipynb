# I. Data Coding and Merging

```python
# Load packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load game logs data
Teams = pd.read_excel("../Data/Game logs 1999-2003.xlsx")
```

```python
# Create a dummy variable for home wins (1 or 0) and visitor wins, and create the â€˜yearâ€™ variable.
Teams['hwin'] = np.where(Teams['home_score'] > Teams['visitor_score'],1,0)
Teams['awin'] = np.where(Teams['home_score'] < Teams['visitor_score'],1,0)
Teams['year'] = Teams['date'].astype(str).str[0:4]
display(Teams)
```

```python
# Create a dataframe that sums the statistics you need by season (year) and by home team. 
# Rename the column for the home team as â€˜Teamâ€™.
Teamshome = Teams.groupby(['home','year'])['visitor_ab', 'visitor_h', 'visitor_2b', 'visitor_3b', 'visitor_hr', 'visitor_sf',
                                           'visitor_hbp','visitor_bb','home_ab', 'home_h', 'home_2b', 'home_3b',
                                           'home_hr', 'home_sf', 'home_hbp', 'home_bb', 'hwin','awin'].sum()
Teamshome = Teamshome.reset_index()
Teamshome = Teamshome.rename(columns = {"home": "Team"}) 
Teamshome
```

```python
# Create a dataframe that sums the statistics you need by season (year) and by visiting team. 
# Rename the column for the visiting team as â€˜Teamâ€™.
Teamsaway = Teams.groupby(['year','visitor'])['visitor_ab', 'visitor_h', 'visitor_2b', 'visitor_3b', 
                                           'visitor_hr', 'visitor_sf', 'visitor_hbp', 'visitor_bb', 
                                           'home_ab', 'home_h', 'home_2b', 'home_3b', 'home_hr', 
                                           'home_sf', 'home_hbp', 'home_bb','hwin','awin'].sum()

Teamsaway = Teamsaway.reset_index()
Teamsaway = Teamsaway.rename(columns = {"visitor": "Team"}) 
Teamsaway
```

```python
# Merge the two dataframes on â€˜yearâ€™ and â€˜Teamâ€™ and create a variable for the total number of wins 
# in the season for each team 
Teams2 = pd.merge(Teamshome, Teamsaway, on =['Team','year'])
pd.set_option('display.max_columns', 50)
display(Teams2)
```

# II. Calculating Performance Metrics

```python
# 1. Create batting average for and batting average against as the sum the relevant statistics for each team as 
# home team and visiting team.
Teams2.columns
```

```python
Teams2['BA_for'] = (Teams2['home_h_x'] + Teams2['visitor_h_y']) / (Teams2['home_ab_x'] + Teams2['visitor_ab_y'])
Teams2['BA_against'] = (Teams2['visitor_h_x'] + Teams2['home_h_y']) / (Teams2['visitor_ab_x'] + Teams2['home_ab_y'])
Teams2
```

```python
# 2. Create the win percentage variable for each team by generating a count of games played and 
#    dividing total wins by total games played.
Teams2['games'] = (Teams2['hwin_x'] + Teams2['hwin_y'] + Teams2['awin_x'] + Teams2['awin_y'])
Teams2['wpc'] = (Teams2['hwin_x']+Teams2['awin_y'])/Teams2['games']
Teams2
```

```python
# 3. Using a similar process to part 2), calculate each teamâ€™s home and away win percentage separately for each season. 
Teams2['hgames'] = Teams2['hwin_x']+Teams2['awin_x']
Teams2['agames'] = Teams2['hwin_y']+Teams2['awin_y']
Teams2['h_wpc'] = Teams2['hwin_x']/Teams2['hgames']
Teams2['a_wpc'] = Teams2['awin_y']/Teams2['agames']
Teams2
```

```python
# 4. Merge each teamâ€™s home and away win percentage for each season into one data frame. 
#    Calculate WPCT diff defined as home win percentage â€“ away win percentage for every team in each season.
Teams2['wpc_diff'] = Teams2['h_wpc'] - Teams2['a_wpc']
Teams2
```

```python
aux = Teams2.loc[Teams2['wpc_diff'].idxmax()]
aux['a_wpc']
```

```python
Teams2['AVG Diff'] = Teams2['BA_for'] - Teams2['BA_against']
min(Teams2['AVG Diff'])
```

# III. Running Regressions

```python
#1. Run an OLS (Ordinary Least Squares) regression of win percentage on batting average for and batting average against.
import statsmodels.formula.api as smf
OLS_BAA = smf.ols(formula = 'wpc ~ BA_for + BA_against', data=Teams2).fit()
OLS_BAA.summary()
```

```python
# 2. Run an OLS (Ordinary Least Squares) regression of win percentage on batting average for and batting average against,
#    constraining the coefficient on each variable to be equal and opposite
Teams2['BA_diff'] = Teams2['BA_for'] - Teams2['BA_against']
OLS_BA_diff = smf.ols(formula = 'wpc ~ BA_diff', data=Teams2).fit()
OLS_BA_diff.summary()
```
